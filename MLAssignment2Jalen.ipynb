```python
import random

#Task 5 (task 4 will come next):
#5. The repository has 4(3 input and 1 output) features and 150 rows, as per instructions in
#step 4 the data you select will be 3(2 input and 1 output) features and 100 rows.
#Here I am selecting petal length and petal width for input (x1 and x2), and Iris setosa and Iris virginica as output in binary form:
#Iris setosa being represented as 0 and Iris virginica as 1
df2 = df[df['species'].isin(['Iris-setosa', 'Iris-virginica'])]
X = df2[['petal_length', 'petal_width']].values.tolist()

#Mapping the species to binary values (0, 1)
y = df2['species'].map({'Iris-setosa': 0, 'Iris-virginica': 1}).tolist()

#Shuffling the data randomly and putting into tuples
#As we can see below, the petal lengths and widths in list form correspond to binaries in 
#list form
data = list(zip(X, y))
random.shuffle(data)
X_shuffled, y_shuffled = zip(*data)
```

```python
#Task 7:
#Split your data in 80:20 (Train : Test) form, train your model with 80 Train data and plot the
#accuracy percentage on train data vs number of epochs.
#Training and plot coming later

#Separating the data to get a 50 50 split. I didn't initially do this and the split wasn't close
#enough to 50%
setosa_data = [(X[i], y[i]) for i in range(len(y)) if y[i] == 0]  # Iris-setosa (class 0)
virginica_data = [(X[i], y[i]) for i in range(len(y)) if y[i] == 1]  # Iris-virginica (class 1)

#shuffling to make it random
random.shuffle(setosa_data)
random.shuffle(virginica_data)

#80% train, 20% test
train_size = 0.8

#Split up Iris-setosa data
setosa_train = setosa_data[:int(train_size * len(setosa_data))]
setosa_test = setosa_data[int(train_size * len(setosa_data)):]

#Split up Iris-virginica data
virginica_train = virginica_data[:int(train_size * len(virginica_data))]
virginica_test = virginica_data[int(train_size * len(virginica_data)):]

# Combine the two for training
train_data = setosa_train + virginica_train

# Combining for test
test_data = setosa_test + virginica_test

random.shuffle(train_data)
random.shuffle(test_data)

X_train, y_train = zip(*train_data)
X_test, y_test = zip(*test_data)

#Completing this task:
#Split the data randomly in 80:20, that is train and test data both should have roughly 50-50%
#data of each class. 

#Printing to show the split. [10 Points]
print(f"Training Set: {sum(1 for y in y_train if y == 0)} Iris-setosa, {sum(1 for y in y_train if y == 1)} Iris-virginica")
print(f"Test Set: {sum(1 for y in y_test if y == 0)} Iris-setosa, {sum(1 for y in y_test if y == 1)} Iris-virginica")
```

```python
#Performing logistic regression for row 1 only, getting the formula down and seeing the pattern
#needed for all rows. Note that this is not task 4 yet. This comes next.

#Initializing weights and prediction for logistic regression
import math

B_0 = 0.0
B_1 = 0.0
B_2 = 0.0

#This function will be used later to save time calculating everytime
def prediction_calc(B0, B1, B2, x1, x2):
    #Computing according to the formula
    output = B0 + B1 * x1 + B2 * x2
    
    #Computing the prediction using sigmoid function
    prediction = 1 / (1 + math.exp(-output))
    
    return prediction 

#saving some time b/c I know the value is 0.5 to start
prediction = 0.5
alpha = 0.3

#Using update rule for row 1
B_0 = B_0 + alpha * (y_train[0] - prediction) * (1-prediction) * 1.0
B_1 = B_1 + alpha * (y_train[0] - prediction) * (1-prediction) * X_train[0][0]
B_2 = B_2 + alpha * (y_train[0] - prediction) * (1-prediction) * X_train[0][1]
prediction = prediction_calc(B_0,B_1,B_2,X_train[0][0],X_train[0][1])
if prediction < 0.5:
    expected_prediction = 0
else:
    expected_prediction = 1
print(f"Prediction = {expected_prediction}, Actual = {y_train[0]}")
```

```python
#The long-awaited Task 4 is done here:
#Make a logistic regression model using two features (for example: “sepal length” and “sepal
#width”) as input and any two flowers (for example “Iris-setosa” and “Iris-virginica”) as binary
#classifiable output.
#Resetting all weights and needed values so I can start fresh for logistic regression of all 100 rows
B_0 = 0.0
B_1 = 0.0
B_2 = 0.0
prediction = 0.5
alpha = 0.3
```

```python
#probably don't need 100 iterations but it didn't hurt. I was able to achieve 
#100% accurracy early on because of the clear separation between the flowers/features

iterations = 100
accuracy_vals = []

#100 epochs
for iteration in range(iterations):
    Correct_predictions = 0

    #looping through all 80 rows, updating weights and prediction accordingly
    for i in range(len(y_train)):
        prediction = prediction_calc(B_0,B_1,B_2,X_train[i][0],X_train[i][1])

        if prediction < 0.5:
            expected_prediction = 0
        else:
            expected_prediction = 1
    
        B_0 = B_0 + alpha * (y_train[i] - prediction) * (1-prediction) * 1.0
        B_1 = B_1 + alpha * (y_train[i] - prediction) * (1-prediction) * X_train[i][0]
        B_2 = B_2 + alpha * (y_train[i] - prediction) * (1-prediction) * X_train[i][1]
    
        if expected_prediction == y_train[i]:
            Correct_predictions += 1
        #Debug: print(f"Iteration {i}: Prediction = {expected_prediction}, Actual = {y_train[i]}")

    #Printing accuracy at the end of each iteration (epoch)
    accuracy = Correct_predictions / len(y_train)
    accuracy_vals.append(accuracy)
    print(f"Iteration {iteration + 1}/{iterations}: Accuracy = {accuracy * 100:.2f}%")


    
        
```

```python
#Plotting for task 7:
#Split your data in 80:20 (Train : Test) form, train your model with 80 Train data and plot the
#accuracy percentage on train data vs number of epochs.

# Plotting accuracy vs iterations
plt.plot(range(1, iterations + 1), accuracy_vals)
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Iteration')
plt.grid(True)
plt.show()
```

```python
#Task 8:
#After the model is trained, test the model with test data and report the accuracy percentage with
#test data.

#Testing with test data
#Resetting the count for correct predictions, and performing logistic
#regression on our test data. Not changing weights obviously.
Correct_predictions = 0
for i in range(len(y_test)):
        prediction = prediction_calc(B_0,B_1,B_2,X_test[i][0],X_test[i][1])

        if prediction < 0.5:
            expected_prediction = 0
        else:
            expected_prediction = 1
    
        if expected_prediction == y_test[i]:
            Correct_predictions += 1
        print(f"Iteration {i}: Prediction = {expected_prediction}, Actual = {y_test[i]}")

#Printing prediction and actual binaries and then printing overall accuracy
accuracy = Correct_predictions / len(y_test)
print(f"Accuracy = {accuracy * 100:.2f}%")
```
